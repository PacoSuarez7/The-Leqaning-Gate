```python
import pandas as pd
import numpy as np

file_path1 = '/content/drive/My Drive/Colab Notebooks/test.csv'
file_path2 = '/content/drive/My Drive/Colab Notebooks/test2.csv'
file_path3 = '/content/drive/My Drive/Colab Notebooks/train.csv'
file_path4 = '/content/drive/My Drive/Colab Notebooks/train2.csv'

datat1 = pd.read_csv(file_path1)
datat2 = pd.read_csv(file_path2)
datat3 = pd.read_csv(file_path3)
datat4 = pd.read_csv(file_path4)
```

```python
#Limpieza y agrupacion de la Data en un solo DataFrame
data_os = pd.concat([datat2, datat4])
data_emp = pd.concat([datat1, datat3])
data_all = data_emp.merge(data_os, how='left', right_on = 'Employee_ID', left_on = 'Employee_ID')
```

```python
data_all[' Computer_OS'].fillna('no registrado', inplace=True)
data_all[' Mobile_OS'].fillna('no registrado', inplace=True)
```

```python
#Validacion de la Data que no contenga NaN
print(data_all[' Computer_OS'].drop_duplicates())
```

```python
print(data_all[' Mobile_OS'].drop_duplicates())
```

Pregunta A: El porcentaje de personas que usan Windows / Mac / Linux / no
registrado como su sistema operativo en su computadora.

```python
#Inicio de solucion a las Preguntas:
r1 = data_all.groupby([' Computer_OS']).count()
r1 = r1['Employee_ID'].apply(lambda x: (x/r1['Employee_ID'].sum()*100).astype(str) + ' %')
```

Pregunta B: El porcentaje de personas que usan iOS / Android / No registrado como su sistema operativo en su dispositivo móvil.

```python
r2 = data_all.groupby([' Mobile_OS']).count()
r2 = r2['Employee_ID'].apply(lambda x: (x/r2['Employee_ID'].sum()*100).astype(str) + ' %')
```

Pregunta C: La edad promedio de la gente que utiliza cada sistema operativo de
escritorio.

```python
r3 = data_all.groupby([' Computer_OS']).mean()
r3 = r3['Age']
```

Pregunta D: El nivel educativo promedio de la gente que utiliza cada sistema
operativo de escritorio.

```python
r4 = data_all.groupby([' Computer_OS']).mean()
r4 = r4['Education_Level']
```

Pregunta E: El número total de tickets generados por la gente con cada sistema
operativo de escritorio.

```python
r5 = data_all.groupby([' Computer_OS']).sum()
r5 = r5[' Computer_tickets']
```
