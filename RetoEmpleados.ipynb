14. Ahora debes hacer la evaluación de las variables para quedarte con las mejores. Calcula la correlación lineal de cada una de las variables con respecto al Attrition.

```python
def color_negative_green(val):
    color = 'green' if val >= 0.1 or val <= -0.1 else 'black'
    return f'color: {color}'

def color_negative_red(val):
    color = 'red' if val > 0.195 or val < -0.195 else 'white'
    return f'color: {color}'

def highlight_value(val):
    if val >= 0.1 or val <= -0.1:
        return 'color: green'
    else:
        return 'color: black'

dep_var = 'Att_Yes'

correlacion = EmpleadosAttrition.corr()
mask = np.triu(np.ones(correlacion.shape), k=1).astype(bool)
corr_df = correlacion.mask(mask)
corr_df = corr_df.where(~np.eye(correlacion.shape[0], dtype=bool))

corr_df = corr_df.style.applymap(color_negative_green, subset=['Att_Yes']).applymap(highlight_value, subset=pd.IndexSlice[[dep_var], :])
corr_df
```
