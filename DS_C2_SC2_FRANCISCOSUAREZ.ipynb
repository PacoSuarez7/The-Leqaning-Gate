# c) Determinar la cantidad de vacunas aplicadas de cada compañía

```python
datat1.groupby('vaccines').count()['country']
```

# d) Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
datat1['total_vaccinations'].sum()
```

#e) Calcular el promedio de vacunas aplicadas por país.

```python
datat1.groupby('country').mean()['total_vaccinations']
```

#f) Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
datat1[datat1['date'] == '2021-01-29']['total_vaccinations'].sum()
```

#g) conDiferencias

```python
conDiferencias = datat1.copy()
conDiferencias['diferecias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias
```

#h) Periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.



```python
min = datat1['date'].min()
max = datat1['date'].max()
max-min
```

#i) conCantidad

```python
conCantidad = datat1.copy()
conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').apply(lambda x: len(x))
conCantidad
```

#j) antes20

```python
antes20 = datat1[datat1['date']<'2020-12-20']
antes20
```

#k) pfizer

```python
pfizer = datat1[datat1['vaccines'].str.contains('pfizer', case=False)]
pfizer
```

#i) Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx

```python
writer = pd.ExcelWriter('/content/drive/My Drive/Colab Notebooks/resultadosReto.xlsx')

# Write each DataFrame to a separate sheet
conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=True)
conCantidad.to_excel(writer, sheet_name='conCantidad', index=True)
antes20.to_excel(writer, sheet_name='antes20', index=True)
pfizer.to_excel(writer, sheet_name='pfizer', index=True)


# Save the Excel file
writer.save()
```
